Existem algumas abordagens para tratamento de dados inconsistentes

-Apagar a coluna --nao é recomendado, pois perderá a variavel.

-Apagar somente os dados com problemas

-Preencher os valores manualmente- é a mais eficiente, mais é inviavel

-Preencher os valores inconsistentes com a média



```python
#Apagar a coluna 
#base.drop('age',1,inplace = True)
#Apagar somente os dados com problemas
#ase.drop(base[base.age<0].index,inplace=True)
#Preencher os valores manualmente
#Preencher com a nédia
```

```python
#Média de todas as colunas
base.mean()
```

```python
base['age'].mean()
```

```python
#devemos desconsiderar os valores inconsistentes
base['age'][base.age>0].mean()
```

```python
#Faremos a substituição dos valores inconsistentes com a média da idade.
base.loc[base.age<0,'age']=base['age'][base.age>0].mean()
```

Devemos aplicar o escalonamento, pois há um intervalo muito grande entre o salario e a idade.
Existem 2 maneiras de fazer este escalonamento:

-Padronizacao: mais recomendado a usar pois ela é robusta, ideal para cenario que tenha outlier


-Normalização


Ambas as formas estão presentes no sklearn, ou seja, não precisa ser implementada.


```python
pd.DataFrame(previsores[:,0:3])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
previsores = scaler.fit_transform(previsores)
```

Neste trecho sklearn já fez a padronização dos dados deixando na mesma escala, assim neste exemplo nao vai considerar a renda(coluna 0) como a mais importante somente pelo fato de terem maiores valores
