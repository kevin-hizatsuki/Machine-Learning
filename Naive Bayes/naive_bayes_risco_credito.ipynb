Uso do Naive Bayes
Vantegem:

-Rapido

-Simplicidade de interpretação

-Trabalha com altas dimensões

-Boas previsões em bases pequenas

Desvantagens

-Combinação de caracteristicas onde cada par de caracteristicas são independentes, ou seja, não faz correlação nenhuma

```python
import pandas as pd
```

```python
base = pd.read_csv('Base de dados/risco-credito.csv')
```

```python
base
```

```python
previsores = base.iloc[:,0:4].values
pd.DataFrame(previsores)
```

```python
classe = base.iloc[:,4].values
pd.DataFrame(classe)
```

```python
from sklearn.preprocessing import LabelEncoder
```

Como GaussianNB não aceita variaveis qualitativos, devemos transformar estes atributos que estão em string para variaveis que representem quantidade, ou seja, numéricos.

```python
labelencoder = LabelEncoder()
previsores[:,0] = labelencoder.fit_transform(previsores[:,0])
previsores[:,1] = labelencoder.fit_transform(previsores[:,1])
previsores[:,2] = labelencoder.fit_transform(previsores[:,2])
previsores[:,3] = labelencoder.fit_transform(previsores[:,3])
```

```python
pd.DataFrame(previsores)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classificador = GaussianNB()
classificador.fit(previsores,classe)
```

Após o treinamento do algoritimo com a base de dados, passaremos um caso de um cliente novo com as seguintes descrições: 

Historia de credito: Boa

Divida: Alta

Garantias: Nenhuma

Renda: >35


```python
resultado = classificador.predict([[0,0,1,2]])
```

Onde o resultado é equivalente à Baixo Risco

```python
pd.DataFrame(resultado)
```

Agora passaremos um caso que disperta a Correção Laplaciana(que seria variaveis que traz resultados zerados):



Historia de credito: Ruim

Divida: Alta

Garantias: Adequada

Renda: <15

```python
resultado = classificador.predict([[0,0,1,2],[3,0,0,0]])
```

```python
pd.DataFrame(resultado)
```

```python
#Não precisamos fazer a transformação das variaveis classe
print(classificador.classes_)
```

```python
print(classificador.class_count_)
```

```python
#São as probabilidade apriori
print(classificador.class_prior_)
```
