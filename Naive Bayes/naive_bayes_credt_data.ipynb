```python
import pandas as pd
```

```python
base = pd.read_csv('Base de dados/credit-data.csv')
```

```python
base
```

```python
#Transformação dos numeros negativos
base.loc[base.age < 0,'age'] = 40.92
base.loc[base.age < 0,'age']
```

```python
previsores = base.iloc[:,1:4].values
classe = base.iloc[:,4].values
```

```python
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer(missing_values = 'NaN',strategy = 'mean',axis = 0)
```

```python
imputer =imputer.fit(previsores[:,1:4])
previsores[:,1:4] = imputer.transform(previsores[:,1:4])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
```

```python
from sklearn.model_selection import train_test_split
```

```python
previsores_treinamento,previsores_teste,classe_treinamento,classe_teste = train_test_split(previsores,classe,test_size = 0.25,random_state = 0)
```

```python
pd.DataFrame(previsores_treinamento)
```

```python
pd.DataFrame(previsores_teste)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classificador = GaussianNB()
```

```python
classificador.fit(previsores_treinamento,classe_treinamento)
```

```python
previsoes = classificador.predict(previsores_teste)
```

```python
pd.DataFrame(previsoes)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

Neste trecho de codigo a sklearn faz a comparação com a base de teste com a base de resposta previsores, e traz uma porcentagem de acerto.

```python
precisao = accuracy_score(classe_teste,previsoes)
```

```python
precisao
```

Com a matriz de confusão conseguimos analisar mellhor a resposta do algoritimo, onde o algoritimo acertou 428 +41 = 469, ou seja, dos 500 dados de treinamento, foi acertado 469 registros,e errou 31.

```python
matriz = confusion_matrix(classe_teste,previsoes)
```

```python
pd.DataFrame(matriz)
```
