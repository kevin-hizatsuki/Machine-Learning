```python
import pandas as pd
```

```python
base = pd.read_csv('Base de dados/census.csv')
```

```python
base
```

```python
previsores = base.iloc[:,0:14].values
classe = base.iloc[:,14].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_previsores = LabelEncoder()
```

```python
labels =labelencoder_previsores.fit_transform(previsores[:,1])
pd.DataFrame(labels)
```

```python
previsores[:,1] =labelencoder_previsores.fit_transform(previsores[:,1])
previsores[:,3] =labelencoder_previsores.fit_transform(previsores[:,3])
previsores[:,5] =labelencoder_previsores.fit_transform(previsores[:,5])
previsores[:,6] =labelencoder_previsores.fit_transform(previsores[:,6])
previsores[:,7] =labelencoder_previsores.fit_transform(previsores[:,7])
previsores[:,8] =labelencoder_previsores.fit_transform(previsores[:,8])
previsores[:,9] =labelencoder_previsores.fit_transform(previsores[:,9])
previsores[:,13] =labelencoder_previsores.fit_transform(previsores[:,13])
```

```python
pd.DataFrame(previsores)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(categorical_features = [1,3,5,6,7,8,9,13]) # Posição das colunas categoricas
previsores = onehotencoder.fit_transform(previsores).toarray()
```

```python
pd.DataFrame(previsores)
```

```python
labelencoder_classe = LabelEncoder()
classe = labelencoder_classe.fit_transform(classe)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
previsores = scaler.fit_transform(previsores)
```

```python
from sklearn.model_selection import train_test_split
```

```python
previsores_treinamento,previsores_teste,classe_treinamento,classe_teste = train_test_split(previsores,classe,test_size = 0.15,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classificador = GaussianNB()
classificador.fit(previsores,classe)
```

```python
previsoes = classificador.predict(previsores_teste)
```

```python
previsoes
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
precisao = accuracy_score(classe_teste,previsoes)
```

```python
precisao
```

Como pode se ver, a taxa de acerto é extremamente baixa, onde um simples random, seria mais eficaz do que este modelo.
